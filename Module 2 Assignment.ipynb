```python
# ANS.3)

def bernoulli_pmf(x, p):
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0.0  

# Example:
print("P(X=1):", bernoulli_pmf(1, 0.7))
print("P(X=0):", bernoulli_pmf(0, 0.7))
print("P(X=2):", bernoulli_pmf(2, 0.7))
```

```python
# ANS.5)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lam = 4      
size = 10000 

data = np.random.poisson(lam=lam, size=size)

plt.figure(figsize=(8, 4))
counts, bins, _ = plt.hist(data, bins=range(0, max(data)+2), density=True, 
                           edgecolor='black', align='left', label='Simulated Data')

x = np.arange(0, max(data)+1)
pmf = poisson.pmf(x, mu=lam)
plt.stem(x, pmf, basefmt=" ", linefmt='r-', markerfmt='ro', label='Poisson PMF')

plt.title(f'Poisson Distribution (λ = {lam})')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.xticks(x)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ANS.9)

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    data = np.array(data)

    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    plt.figure(figsize=(10, 4))
    plt.plot(z_scores, marker='o', linestyle='-', color='blue', label='Z-scores')
    plt.axhline(y=0, color='gray', linestyle='--')
    plt.axhline(y=1, color='green', linestyle='--', linewidth=0.8)
    plt.axhline(y=-1, color='green', linestyle='--', linewidth=0.8)
    plt.axhline(y=2, color='red', linestyle='--', linewidth=0.8)
    plt.axhline(y=-2, color='red', linestyle='--', linewidth=0.8)
    plt.title("Z-Scores of Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-Score")
    plt.grid(True)
    plt.legend()
    plt.show()

    return z_scores

# Example
sample_data = np.random.normal(loc=50, scale=10, size=100)
z = calculate_and_plot_zscores(sample_data)
```

```python
# ANS.10)

import numpy as np
import matplotlib.pyplot as plt

population_dist = np.random.exponential(scale=2.0, size=100000)  # highly skewed distribution
sample_sizes = [5, 30, 100]  
num_samples = 1000           

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(population_dist, bins=50, color='orange', edgecolor='black', density=True)
plt.title("Original Exponential Distribution (Population)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)

for i, n in enumerate(sample_sizes, start=2):
    sample_means = []

    for _ in range(num_samples):
        sample = np.random.choice(population_dist, size=n, replace=True)
        sample_means.append(np.mean(sample))

    plt.subplot(2, 2, i)
    plt.hist(sample_means, bins=40, color='skyblue', edgecolor='black', density=True)
    plt.title(f"Distribution of Sample Means (n={n})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# ANS.12)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():
    """
    Plots the Standard Normal Distribution with mean = 0 and std = 1
    """
    x = np.linspace(-4, 4, 1000)  # Z-scores from -4 to 4

    y = norm.pdf(x, loc=0, scale=1)

    plt.figure(figsize=(8, 4))
    plt.plot(x, y, color='blue', label='Standard Normal PDF')
    plt.fill_between(x, y, color='skyblue', alpha=0.4)
    
    plt.axvline(x=0, color='black', linestyle='--', label='Mean (μ = 0)')
    plt.axvline(x=1, color='green', linestyle='--', alpha=0.6, label='σ = 1')
    plt.axvline(x=-1, color='green', linestyle='--', alpha=0.6)

    plt.title("Standard Normal Distribution (μ = 0, σ = 1)")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_standard_normal_distribution()
```

```python
# ANS.14)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """
    Calculate Z-score of a given value.
    """
    return (x - mean) / std

mean = 100
std = 15
x_value = 130 

z = calculate_z_score(x_value, mean, std)
print(f"Z-score for x = {x_value} is {z:.2f}")

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 4))
plt.plot(x, y, label="Standard Normal Distribution", color='blue')
plt.fill_between(x, y, where=(x <= z), color='skyblue', alpha=0.5, label='Area ≤ Z')

plt.axvline(z, color='red', linestyle='--', label=f"Z = {z:.2f}")
plt.title(f"Z-score Visualization (x = {x_value}, μ = {mean}, σ = {std})")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ANS.15)

import numpy as np
from scipy.stats import norm

mu_population = 50   
sigma = 5             
n = 40                
x_bar = 52           
alpha = 0.05         

z_score = (x_bar - mu_population) / (sigma / np.sqrt(n))
print(f"Z-score: {z_score:.2f}")

z_critical = norm.ppf(1 - alpha/2)
print(f"Critical Z-value (two-tailed): ±{z_critical:.2f}")

if abs(z_score) > z_critical:
    print("Result: Reject the null hypothesis (significant difference)")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference)")
```

```python
# ANS.16)

import numpy as np
from scipy import stats

data = [78, 85, 92, 88, 76, 81, 95, 89, 84, 90]

n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1) 

confidence = 0.95
alpha = 1 - confidence

t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
margin_of_error = t_critical * (std / np.sqrt(n))

lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Sample Mean: {mean:.2f}")
print(f"{confidence*100:.0f}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
# ANS.17)

import numpy as np
from scipy import stats

np.random.seed(42) 
mu, sigma = 100, 15 
n = 50 

data = np.random.normal(loc=mu, scale=sigma, size=n)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

confidence = 0.95
alpha = 1 - confidence
t_critical = stats.t.ppf(1 - alpha / 2, df=n-1) 

margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# ANS.21)

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
dataset1 = np.random.normal(loc=50, scale=5, size=1000)   # Mean = 50, Std = 5
dataset2 = np.random.normal(loc=50, scale=15, size=1000)  # Mean = 50, Std = 15

std1 = np.std(dataset1)
std2 = np.std(dataset2)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

datasets = ['Dataset 1', 'Dataset 2']
std_values = [std1, std2]

plt.figure(figsize=(6, 4))
bars = plt.bar(datasets, std_values, color=['blue', 'green'])
plt.title('Comparison of Standard Deviations')
plt.ylabel('Standard Deviation')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{height:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# ANS.23)

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a dataset from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10

# Step 2: Apply Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

print(f"Original Mean: {mean:.2f}, Std Dev: {std:.2f}")
print(f"Normalized Mean: {np.mean(z_scores):.2f}, Std Dev: {np.std(z_scores):.2f}")

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='salmon', edgecolor='black')
plt.title('Z-Score Normalized Distribution')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# ANS.24)

import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt

def analyze_normal_distribution_skew_kurtosis(mean=0, std=1, size=1000):
   
    data = np.random.normal(loc=mean, scale=std, size=size)

    skewness = skew(data)
    kurt = kurtosis(data)

    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (Fisher's): {kurt:.4f} (0 = normal)")

    plt.figure(figsize=(8, 4))
    plt.hist(data, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Normally Distributed Data')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

analyze_normal_distribution_skew_kurtosis()
```
